```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data.info()
```

```python
data.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
data['Gender'].value_counts()
```

```python
train = data.dropna(how = 'any', axis = 0)
```

```python
list(train.columns)
```

```python
train.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
X_train = train.iloc[:, 1: 12].values
Y_train = train.iloc[:, 12].values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.preprocessing import LabelEncoder


X = X_train
y = Y_train
feature_names = ['Gender',
 'Married',
 'Dependents',
 'Education',
 'Self_Employed',
 'ApplicantIncome',
 'CoapplicantIncome',
 'LoanAmount',
 'Loan_Amount_Term',
 'Credit_History',
 'Property_Area']

le = LabelEncoder()

for col in range(X.shape[1]):
    if isinstance(X[0, col], str):
        X[:, col] = le.fit_transform(X[:, col].astype(str))

if isinstance(y[0], str):
    y = le.fit_transform(y.astype(str))


clf = RandomForestClassifier(n_estimators = 160, criterion = 'gini', max_depth = 7, random_state = 42)
clf.fit(X, y)
```

```python
test = pd.read_csv("test.csv")
test.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
test_data = test.dropna(how = 'any', axis = 0)
test_data.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
females = test_data[test_data['Gender'] == "Female"]
males = test_data[test_data['Gender'] == "Male"]
```

```python
X_men_test = males.iloc[:, 1:12].values
X_women_test = females.iloc[:, 1:12].values
Y_men_test = males.iloc[:, 11].values
Y_women_test = females.iloc[:, 11].values
```

```python

le = LabelEncoder()

for col in range(X_men_test.shape[1]):
    if isinstance(X_men_test[0, col], str):
        X_men_test[:, col] = le.fit_transform(X_men_test[:, col].astype(str))

for col in range(X_women_test.shape[1]):
    if isinstance(X_women_test[0, col], str):
        X_women_test[:, col] = le.fit_transform(X_women_test[:, col].astype(str))


if isinstance(Y_men_test[0], str):
    Y_men_test = le.fit_transform(Y_men_test.astype(str))

if isinstance(Y_women_test[0], str):
    Y_women_test = le.fit_transform(Y_women_test.astype(str))
```

```python
y_pred = clf.predict(X_women_test)
y_pred


from sklearn import metrics
print('The accuracy of Random Forest Classification is: ', metrics.accuracy_score(y_pred, Y_women_test))
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn import metrics
%matplotlib inline


param_grid = {
    'n_estimators': [100, 150, 200],
    'max_depth': [5, 7, 9, None],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=5,
                           n_jobs=-1,
                           scoring='accuracy')

grid_search.fit(X, y)

print("Best parameters found: ", grid_search.best_params_)
print("Best accuracy found: ", grid_search.best_score_)

best_clf = grid_search.best_estimator_



scaler = StandardScaler()
X = scaler.fit_transform(X)
X_women_test = scaler.transform(X_women_test)

cv_scores = cross_val_score(best_clf, X, y, cv=5, scoring='accuracy')
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", np.mean(cv_scores))



best_clf.fit(X, y)
y_pred = best_clf.predict(X_women_test)
print('The accuracy of Random Forest Classification is: ', metrics.accuracy_score(y_pred, Y_women_test))
```
